## Setup

```python
from polis_client import PolisClient
from pprint import pprint

polis = PolisClient()
```

## Interacting with the default Pol.is server

```python
comments = polis.get_comments(conversation_id="2demo")
if isinstance(comments, list) and len(comments) > 0:
    pprint(comments[0].to_dict())
```

```python
report = polis.get_report(report_id="r2dfw8eambusb8buvecjt")
if report:
    pprint(report.to_dict())
```

```python
conversation = polis.get_conversation(conversation_id="2demo")
if conversation and not isinstance(conversation, str):
    pprint(conversation.to_dict())
```

```python
math = polis.get_math(conversation_id="2demo")
pprint(math.to_dict(), depth=1)
```

```python
votes = polis.get_votes(conversation_id="2demo", pid=2)
if isinstance(votes, list) and len(votes) > 0:
    pprint([v.to_dict() for v in votes])
```

## Interacting with a Polis fork

```python
voxit = PolisClient(base_url="https://voxit.sitra.fi/api/v3")

conversation_voxit = voxit.get_conversation(conversation_id="2asfdaahhy")
pprint(conversation_voxit.to_dict())
```
